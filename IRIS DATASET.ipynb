```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("iris.xls")          #to read dataset into python
print(data)                             #to display the dataset
```

```python
print(data.size)                         #to display the size of dataset
print(data.shape)                        #to display shape(no of rows and columns) of a dataset
```

```python
print(data.columns)                      #to display column names of dataset
```

```python
data[['SL','SW','PL','PW']].mean()       #to display the mean of each columns in a datase
```

```python
data.isna().sum()                        #to check for null values in a dataset
```

```python
df=pd.DataFrame(data)                    #to convert dataset to dataframe
df
```

```python
sns.countplot(x=data.Classification)
plt.show()
```

```python
sns.scatterplot(x='SL',y='SW',data=data,hue='Classification')
plt.show()
```

```python
plt.scatter(data['PL'],data['PW'])
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```
